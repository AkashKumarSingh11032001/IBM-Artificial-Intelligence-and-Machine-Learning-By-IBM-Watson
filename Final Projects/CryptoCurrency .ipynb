# CryptoCurrency Analysis  

## Importing Bitcoin Data

```python
import pandas as pd
from datetime import datetime
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%d-%m-%Y')
btc=pd.read_csv('Dataset/Bitcoin.csv',parse_dates=['Date'],index_col='Date', date_parser=dateparse)
btc = btc.iloc[::-1]
btc.tail(5)
```

```python
maxValue=btc[btc['Close']==max(btc.Close)]
print("Highiest value of bitcoin")
maxValue
```

```python
btc.describe()
```

```python
btc.info()
```
