```python
# converting each column to 'category' type !!!
data = data.astype('category')
data.dtypes
```

```python
# now converting Catogerial to Ordinal value !!!

from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
    
data
```

```python
# dropping "veil-type" as it not contributing in analysis as all are '0'

data = data.drop(['veil-type'],axis = 1)
```

```python
data
```

```python
# spliting data into ddependent and independent

x = data.iloc[:,1:23]
y = data.iloc[:,0]
```

```python
# splitting trainnig dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x , y , test_size = .20, random_state = 1)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(criterion = "entropy",max_depth=10)
decision_tree.fit(x_train, y_train)
predictValues =decision_tree.predict(x_test)




```

```python
from sklearn.metrics import accuracy_score
import numpy as np

print(accuracy_score(y_test, predictValues))
```
