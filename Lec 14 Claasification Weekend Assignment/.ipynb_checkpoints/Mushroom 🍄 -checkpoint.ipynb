```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Dataset/mushroom.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data['m_class'].value_counts()
```

```python
# converting each column to 'category' type !!!
data = data.astype('category')
data.dtypes
```

```python
# now converting Catogerial to Ordinal value !!!

from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
    
data
```

```python
# dropping "veil-type" as it not contributing in analysis as all are '0'

data = data.drop(['veil-type'],axis = 1)
```

```python
data
```

```python
# spliting data into ddependent and independent

x = data.iloc[:,1:23]
y = data.iloc[:,0]
```

```python
# splitting trainnig dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x , y , test_size = .20, random_state = 1)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

## Using ~ Decission Tree 

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(criterion = "entropy",max_depth=10)
decision_tree.fit(x_train, y_train)



```

```python
predict =decision_tree.predict(x_test)


from sklearn.metrics import accuracy_score
import numpy as np

print(accuracy_score(y_test, predict))
```

## Using Logistic Regression

```python
# applying Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression ()
lr.fit (x_train, y_train)
```

```python
# accuracy check
predict = lr.predict (x_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predict))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix (y_test, predict)

print(cm)
```
